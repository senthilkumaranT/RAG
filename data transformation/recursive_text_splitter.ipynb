```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader

```

```python
doc = PyPDFLoader("grossman98-Data-minin-research-opportunities.pdf").load()
type(doc)
```

```python
text= RecursiveCharacterTextSplitter("\n\n",chunk_size=100,chunk_overlap=50)
tex=text.split_documents(doc)
tex
```

```python
text= RecursiveCharacterTextSplitter("\n",chunk_size=100,chunk_overlap=50)
tex=text.split_documents(doc)
tex
```

```python
text= RecursiveCharacterTextSplitter(" ",chunk_size=100,chunk_overlap=50)
tex=text.split_documents(doc)
tex
```

```python
text= RecursiveCharacterTextSplitter("",chunk_size=100,chunk_overlap=50)
tex=text.split_documents(doc)
tex
```
