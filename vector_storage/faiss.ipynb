fassbook AI similarity search (FAISS)

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

data ingestion 


```python
documents = TextLoader("speech.txt",encoding="utf-8").load()

text_1 = CharacterTextSplitter(chunk_size=1000 , chunk_overlap =30)
text = text_1.split_documents(documents)
text
```

```python
embeddings = OllamaEmbeddings(model= "gemma:2b")
db=FAISS.from_documents(text,embeddings)
```

```python
db
```

```python
query1 = "summarize the content "
docs = db.similarity_search(query1)
docs
```

```python
## retriever

retriver=db.as_retriever()
docs = retriver.invoke(query1)
```

```python
docs_and_score = db.similarity_search_with_score(query1)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query1)
embedding_vector
```

```python
doc_score=db.similarity_search_with_score(embedding_vector)
doc_score
```

```python
### saving and loading 

db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",embeddings,allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query1)
```
