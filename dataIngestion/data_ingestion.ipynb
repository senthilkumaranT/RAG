```python
## wb based loader 

from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_path=("https://sethu.ac.in/"),bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("post-title","post-content","post-header"))))

loader
```

```python
tex =loader.load()
tex
```
